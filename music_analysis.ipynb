visualization of data , 
plotting graphs for analyzing data

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = ['danceability', 'energy', 'speechiness', 'acousticness', 
            'instrumentalness', 'valence', 'tempo', 'loudness']

for feature in features:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[features].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Audio Features")
plt.show()
```

```python
sns.boxplot(data=df, x='explicit', y='valence')
plt.title("Valence (Happiness) of Explicit vs Non-Explicit Songs")
```

```python
df['key'].value_counts().head(10).plot(kind='bar', title='Most Common Musical Keys')
```

```python
# Group by year and count the number of songs
songs_per_year = df['year'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(14, 6))
sns.barplot(x=songs_per_year.index, y=songs_per_year.values, palette='viridis')
plt.title('Number of Songs Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['mood'] = pd.cut(df['valence'], bins=[0, 0.3, 0.7, 1.0], labels=['Sad', 'Neutral', 'Happy'])
df['mood'].value_counts()
plt.figure(figsize=(6,4))
sns.countplot(x='mood', data=df, palette='pastel')
plt.title('Mood Distribution Based on Valence')
plt.xlabel('Mood')
plt.ylabel('Number of Songs')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

displaying top 10 artists who released highest number of songs

```python

top_artists = df['artists'].value_counts().head(10)

# Display
print(top_artists)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='coolwarm')
plt.title("Top 10 Artists by Number of Songs")
plt.xlabel("Number of Songs")
plt.ylabel("Artist")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

analysing top 10 songs according to its final score that is according to its features

```python

df['final_score'] = (df['energy'] + df['danceability'] + df['valence']) / 3

top_10 = df.sort_values(by='final_score', ascending=False).head(10)

top_10[['name', 'artists', 'final_score']]
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='final_score', y='name', data=top_10, palette='Blues_r')
plt.title("Top 10 Songs by Combined Feature Score")
plt.xlabel("Score")
plt.ylabel("Song Name")
plt.tight_layout()
plt.show()
```

top 10 sad songs

```python
# Filter and sort songs with lowest valence
top_10_sad = df.sort_values(by='valence', ascending=True).head(10)

# Display
top_10_sad[['name', 'artists', 'valence']]
```

top 10 happy songs

```python
# Filter and sort songs with highest valence
top_10_happy = df.sort_values(by='valence', ascending=False).head(10)

# Display
top_10_happy[['name', 'artists', 'valence']]
```

```python
# Combine all results into one dictionary 
top_10_sad['Mood'] = 'Sad'
top_10_happy['Mood'] = 'Happy'

combined = pd.concat([top_10_sad, top_10_happy])
combined[['name', 'artists', 'Mood', 'valence', 'acousticness']].reset_index(drop=True)
```
